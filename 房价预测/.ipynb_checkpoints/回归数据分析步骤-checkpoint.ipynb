## 思路

- Understand the problem. We'll look at each variable and do a philosophical analysis about their meaning and importance for this problem.  
- Univariable study. We'll just focus on the dependent variable ('SalePrice') and try to know a little bit more about it.  
- Multivariate study. We'll try to understand how the dependent variable and independent variables relate.  
- Basic cleaning. We'll clean the dataset and handle the missing data, outliers and categorical variables.  
- Test assumptions. We'll check if our data meets the assumptions required by most multivariate techniques.

## 2. First things first: analysing 'SalePrice

```python
# 探索我们预测目标(房价情况)
df_train.SalePrice.describe()
```

```python
# 查看目标的分布情况
sns.distplot(df_train.SalePrice)
```

结论：

（２）有峰值

```python
# 计算锋度和偏度
print("skew:%f" % (df_train.SalePrice.skew()))
print("kurt:%f" % (df_train.SalePrice.kurt()))
```

## 'SalePrice', her buddies and her interests

### 数字性变量

```python
## 探索房屋居住面积
var = "GrLivArea"
data = pd.concat([df_train.SalePrice, df_train[var]], axis=1)
data.plot.scatter(x=var, y="SalePrice", ylim = (0, 800000))
```

结论：  
房屋面积与房价呈现正相关

```python
## TotalBsmtSF 地下室面积
var = "TotalBsmtSF"
data = pd.concat([df_train.SalePrice, df_train[var]], axis=1)
data.plot.scatter(x=var, y="SalePrice", ylim = (0, 800000))
```

结论：  
(1)存在地下面积为零却也有不错房价的房屋，但是整体上与房价有相关性质

### 分类性变量

```python
## 房屋的整体质量
var = "OverallQual"
data = pd.concat([df_train.SalePrice, df_train[var]], axis=1)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(x=var, y="SalePrice", data=data, ax=ax)
ax.axis(ymin = 0, ymax=800000);
```

结论：  
(1) 房屋的整体质量越好，则其房价也会越好。

```python
## "YearBuilt" 房屋修建年限
var = "YearBuilt"
data = pd.concat([df_train.SalePrice, df_train[var]], axis=1)
fig, ax = plt.subplots(figsize=(24, 6))
sns.boxplot(x=var, y="SalePrice", data=data, ax=ax)
ax.axis(ymin = 0, ymax=800000);
plt.xticks(rotation = 90);
```

结论：  
（１）整体表现买新不买旧

## Keep calm and work smart

```python
## 相关性矩阵，分析特征之间的相关性
corrmat = df_train.corr()
fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(corrmat, square=True, vmax=0.8)
```

结论：  
（１）特征之间的多重共线性

```python
## 缩小范围
k = 10 # 取出与房价相关系数最高的10个特征
cols = corrmat.nlargest(10, "SalePrice")["SalePrice"].index # 取得特征名称
cols.values
```

```python
cm = np.corrcoef(df_train[cols].values.T) # 每一行表示一个变量
sns.set(font_scale=1.25) # 标签字体大小

hm = sns.heatmap(cm
                 , annot=True
                 , fmt="0.2f"
                 , square=True # 是否正
                 , annot_kws={"size":10}
                 , yticklabels=cols.values
                 , xticklabels=cols.values
                )
```

结论：  
(1)OverallQual、HrLivArea和TotalBsmtSF与房价有较强的关系  
(2)Garage*之间存在多重共性性,所以，二选一，选择了GarageCars(与房价有较强的相关性)，此外，stFlrSF和TotalBsmtSF也做同样的处理，选择TotalBsmtSF，还有GrliveArea和TotRmsAbvGrd

```python
cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars'
        , 'TotalBsmtSF', 'FullBath', 'YearBuilt']

df_train[cols].info()
```

```python
## 查看消除共线的数据的关系, 查看各个特征之间的相关性
sns.set()
sns.pairplot(df_train[cols], height=2.5);
```

结论：  
(1)通过此时绘制相关关系图，可以看出现在特征之间不存在多重共线性

## 4. Missing data
- How prevalent is the missing data?  
- Is missing data random or does it have a pattern?

```python
# 查看缺失情况
total = df_train.isnull().sum().sort_values(ascending = False)  # 减序排列
total
```

```python
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending=False)
```

```python
(df_train.isnull().sum()/len(df_train)).sort_values(ascending=False)
```

```python
missing_data = pd.concat([total, percent], axis=1)
missing_data.columns = ["total", "percent"]
missing_data.head(20)
```

```python
missing_data[missing_data.total>1].index
```

```python
#＃　处理缺失值
df_train = df_train.drop(missing_data[missing_data.total>1].index, 1)
```

```python
# 现在无缺失值
df_train.info()
```

### 异常值 

```python
df_train.describe().T
```

```python
## 数据标准化来处理数据的异常值(必须是二维数据)
saleprice = StandardScaler().fit_transform(df_train.SalePrice[:, np.newaxis])
```

```python
low_range = saleprice[saleprice[:, 0].argsort()[:10]]
hight_range = saleprice[saleprice[:, 0].argsort()[-10:]]
print("Low_range:\n", low_range)
print("Hight_range:\n", hight_range)
```

#### 双变量分析

```python
## 变量GrLivArea
var = 'GrLivArea'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
```

结论：双变量分析可得，右下角出现了两个离群点，它对于整体的数据不具有影响，可以作为异常值删除，此外，虽然最上面也出现了两个孤立的点，但是，其数据趋势与整体相仿，所以不作为异常值处理。

```python
# 删除GrLivArea异常值(右下角的离群点)
df_train.sort_values(by="GrLivArea", ascending=False)[:2]
```

## 5. Getting hard core

```python
## 数据分布的正态性
sns.distplot(df_train["SalePrice"], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'], plot=plt)
```

```python
#　applying log transformation
df_train['SalePrice'] = np.log(df_train['SalePrice'])
```

```python
# 数据变正态分布
sns.distplot(df_train["SalePrice"], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'], plot=plt)
```

```python
##  GrLivArea的数据分布
sns.distplot(df_train['GrLivArea'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['GrLivArea'], plot=plt)
```

```python
## 数据偏态
df_train["GrLivArea"] = np.log(df_train.GarageArea)
```

```python
# 地下室面积
sns.distplot(df_train['TotalBsmtSF'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['TotalBsmtSF'], plot=plt)
```

```python
## log 转换出现的一些问题，消除除零错误
df_train["HasBsmt"] = pd.Series(len(df_train.TotalBsmtSF), index=df_train.index)
```
