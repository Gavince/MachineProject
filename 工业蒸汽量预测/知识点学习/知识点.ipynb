```python
import numpy as np
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = load_iris()
```

```python
X = data.data
```

```python
plt.scatter(x=X[:, 1],y=[X[:, 0]])
```

```python
sns.kdeplot(X[:, 1], shade=True)
```

```python
sns.violinplot(data.target, y=X[:, 1])
```

```python
sns.boxplot(X[:,1], orient="v")
```
